```python
def find_all_indices(string_to_search, term_to_find):
    lower_string = string_to_search.lower()
    lower_term = term_to_find.lower()

    len_string = len(lower_string)
    len_term = len(lower_term)

    found_indices = []

    for i in range(len_string - len_term + 1):
        if lower_string[i:i + len_term] == lower_term:
            found_indices.append(i)

    return found_indices

input_string = input("Enter a string to search through: ")

indices = find_all_indices(input_string, "the")

if indices:
    print(f"\nThe substring 'the' was found at the following starting indices:")
    print(indices)
else:
    print(f"\nThe substring 'the' was not found in your string.")

```
