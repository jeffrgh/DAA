```python
def compute_lps(pattern):
    n = len(pattern)
    lps = [0] * n
    length = 0
    i = 1

    while i < n:
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1

    return lps


input_string = "ABCBCABC"
lps_table = compute_lps(input_string)

print(f"The string is: {input_string}")
print(f"The LPS Table is: {lps_table}")
```
