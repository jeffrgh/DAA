```python
def kmp_search(main_string, pattern):
    n = len(main_string)
    m = len(pattern)

    lps = compute_lps_array(pattern)
    print(f"Computed LPS array for '{pattern}': {lps}")

    i = 0
    j = 0
    found_indices = []

    while i < n:
        if pattern[j] == main_string[i]:
            i += 1
            j += 1

        if j == m:
            found_index = i - j
            found_indices.append(found_index)
            print(f"\nPattern '{pattern}' found at index: {found_index}")
            j = lps[j - 1]

        elif i < n and pattern[j] != main_string[i]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1

    if not found_indices:
        print(f"\nPattern '{pattern}' not found in the main string.")

    return found_indices

def compute_lps_array(pattern):
    m = len(pattern)
    lps = [0] * m
    length = 0
    i = 1

    while i < m:
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1
    return lps

if __name__ == "__main__":
    text = "CATSABCBCABCDOGSABCBCABC"
    pattern = "ABCBCABC"

    print(f"Main String: {text}")
    print(f"Pattern    : {pattern}\n")

    kmp_search(text, pattern)

```
