```python
def create_and_print_matrix(size):
    matrix = []
    current_number = 1
    print("\nGenerated Matrix:")
    for i in range(size):
        row = []
        for j in range(size):
            row.append(current_number)
            current_number += 1
        print(" ".join(f"{num:<3}" for num in row))
    print()
    return matrix

def main():
    try:
        n = int(input("Enter the size (n) for the n*n matrix: "))

        if n <= 0:
            print("Please enter a positive number for the matrix size.")
            return

        matrix = create_and_print_matrix(n)

        i = int(input(f"Enter the row index 'i' (0 to {n-1}): "))
        j = int(input(f"Enter the column index 'j' (0 to {n-1}): "))

        if 0 <= i < n and 0 <= j < n:
            value = (i * n) + j + 1
            print(f"\nThe value at row {i} and column {j} is: {value}")
        else:
            print("\nError: Index out of bounds.")
            print(f"Please make sure your row and column are between 0 and {n-1}.")

    except ValueError:
        print("\nError: Invalid input. Please enter integers only.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```
