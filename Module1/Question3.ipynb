```python
import csv

def find_letters_in_csv(filename="students.csv"):
    found_locations = []
    try:
        with open(filename, mode='r', newline='', encoding='utf-8') as csvfile:
            csv_reader = csv.reader(csvfile)

            row_number = 0
            for row in csv_reader:
                row_number += 1
                col_number = 0
                for cell in row:
                    col_number += 1
                    lower_cell = cell.lower()
                    if 'a' in lower_cell and 'i' in lower_cell:
                        location = f"Row {row_number}, Column {col_number}"
                        print(f"Found 'a' and 'i' in cell at: {location} (Value: '{cell}')")
                        found_locations.append(location)

        if not found_locations:
            print(f"No cells containing both 'a' and 'i' were found in '{filename}'.")

    except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found. Please make sure it's in the same directory.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    print("Searching for cells containing both 'a' and 'i' in students.csv...")
    find_letters_in_csv()

```
